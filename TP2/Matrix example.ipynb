```python
import sympy as sym
import numpy as np
import cmath
from cmath import sqrt
import math as m

def polar_to_rectangular(r, theta):
    x = r * m.cos(m.radians(theta))  # Convertimos theta a radianes
    y = r * m.sin(m.radians(theta))
    return x, y

# Ejemplo de uso
r1= 0.5566  # Magnitud
t1= 169.1  # Ángulo en grados

r2= 7.249
t2= 77.8

r3= 0.047
t3= 52.8

r4= 0.0975
t4= -64.4

pol=[
        [r1, t1],
        [r2, t2],
        [r3, t3],
        [r4, t4]
]

s=[
    # [x, y]
      [0, 0], #S11, codifica en S[0, 0]
      [0, 0], #S12, codifica en S[0, 1]
      [0, 0], #S21, codifica en S[1, 0]
      [0, 0]  #S22, codifica en S[1, 1]
]

for i in range(3):
    for j in range(3):
        valores= polar_to_rectangular(pol[i][j], pol[i][j+1])
        for k= 0 in range(1):
            s[i][k]= valores[k]

# Imprimo la matriz.
print('La matriz es la siguiente: ')
for row in s:
    for column in row:
        print(column, end= " ")
    print()

# s11= complex(r[0], t[0])
# s21= complex(r[1], t[1])
# s12= complex(r[1], t[1])
# s22= complex(r[1], t[1])
```
