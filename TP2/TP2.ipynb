## Cálculo de Factor de Rollet $K$

```python
import sympy as sym
import numpy as np
import cmath
from cmath import sqrt
import math as m


#Variables para la fórmula del COEFICIENTE DE ENTRADA.
f, K, B1, B2, C1, C2, dt, zs, zl = sym.symbols('f, K, B_{1}, B_{2}, C_{1}, C_{2}, \Triangle, zs, zl')


# f=1.8*10**9
f= 1.6*10**9

#Ejemplo de clase
# s11= complex (-0.5482896290575852, 0.3601589686067746)
# s21= complex (0.08459502304964918, 0.0879072356250003)
# s12= complex (1.6348387899078822, 1.617807816464159)
# s22= complex (-0.07862084345409541, -0.15297961620610315)

# #Ejemplo EA3_TP2_Beamer.pdf
# s11= complex (-0.6995217942614798, 0.32916997030895756)
# s21= complex (1.7174245878476282, 2.75915798479471)
# s12= complex (0.045225161461749014, 0.05486250787886689)
# s22= complex (-0.5655334102527199, 0.1903230198581583)

# Caso 1.6GHz, Vce= 1.0V     Ic=37mA
s11= complex(-0.6069, 0.1424)
s12= complex(0.0290, 0.0375)
s21= complex(1.2071, 5.4448)
s22= complex(0.0348, -0.0550)

dt= s11*s22 - s12*s21
abs_dt= abs(dt)
ang_dt_deg= cmath.phase(dt)*(180/np.pi)

K= (1- abs(s11)**2 - abs(s22)**2 + abs(dt)**2)/(2*abs(s12*s21))

if K>1 and abs_dt<1:
    print("El transistor en este punto es INCODICIONALMENTE ESTABLE")
else:
    print("El transistor en este punto es CONDICIONALMENTE ESTABLE")

print("\nCálculo del FACTOR DE ROLLET")
print("K= ", K)
print("Cálculo de DELTA")
print("Valor de dt= ", dt)
print("Valor abs(dt)= ", abs_dt)
print("Ángulo de dt= ", ang_dt_deg)
```

```python
def faseDeg(complejo):
    #Retorno ángulo en radianes
    return cmath.phase(complejo)*(180/np.pi) 

B1= 1+ abs(s11)**2 - abs(s22)**2 - abs(dt)**2
B2= 1+ abs(s22)**2 - abs(s11)**2 - abs(dt)**2

C1= s11-(dt*(s22.conjugate()))
C2= s22-(dt*(s11.conjugate()))

C1_m= abs(C1)
C2_m= abs(C2)

C1_ang= faseDeg(C1)
C2_ang= faseDeg(C2)

print("Condicion maxima ganancia")

print("B1=",B1)
print("B2=",B2)
print("C1=",C1)
print("C2=",C2)
print("|C1|= ", C1_m)
print("|C2|= ", C2_m)
print("C1_ang= ", faseDeg(C1))
print("C1_ang= ", faseDeg(C2))
print("\n\nValores de Gamma_in:")

gamma_in_m= abs((B1 - ((B1)**2 - 4*(abs(C1))**2)**(0.5) )/(2*abs(C1)))

print("|gamma_in|=",gamma_in_m)

#now we calculate the angle of "gamma_in_ang", is equal to the angle of C1 in degrees
gamma_in_ang= np.rad2deg(cmath.phase(C1))
print("gamma_in_ang=",gamma_in_ang)
print("\nValores de Gamma_out:")
gamma_out_m= abs((B2 - ((B2)**2 - 4*(abs(C2))**2)**(0.5) )/(2*abs(C2)))

print("|gamma_out|=",gamma_out_m)

#now we calculate the angle of "gamma_out_ang", is equal to the angle of C2 in degrees
gamma_out_ang= np.rad2deg(cmath.phase(C2))
print("gamma_out_ang=",gamma_out_ang)

print("\nGamma_in y Gamma_out en complejo")
#declarar gamma_in y gamma_out como numeros complejos
gamma_in= cmath.rect(gamma_in_m, np.deg2rad(gamma_in_ang))
gamma_out= cmath.rect(gamma_out_m, np.deg2rad(gamma_out_ang))

print("gamma_in=",gamma_in)
print("gamma_out=",gamma_out)
```

```python
print("Realizo los conjugados delos Gamma_in y Gamma_out.")
#gamma_s es el coeficiente de reflexión en la carga
gamma_s= gamma_in.conjugate()
gamma_l= gamma_out.conjugate()

#los modulos de gamma_s y gamma_l 

gamma_s_m= abs(gamma_s)
gamma_l_m= abs(gamma_l)



print("gamma_s=",gamma_s)
print("gamma_l=",gamma_l)

print("|gamma_s|=",gamma_s_m)
print("|gamma_l|=",gamma_l_m)

print("\n\n")
if(K>=1, gamma_in_m<=1, gamma_out_m<=1, gamma_s_m<=1, gamma_l_m<=1):
    print("ESTABLE")
else:
    print("INESTABLE")
print("\n\n")


zin= 50*(1+gamma_in)/(1-gamma_in)
zout= 50*(1+gamma_out)/(1-gamma_out)

print("zin=",zin)
print("zout=",zout)
```

```python
print("Las impedancias conjugadas de Zo_in y Zo_out son las siguientes")
#zs es el conjugado de zin
zs= zin.conjugate()

#zl es el conjugado de zout
zl= zout.conjugate()

print("Zs=",zs)
print("Zl=",zl)


#Extrae la parte real e imaginaria de z_in

Rs_in= zin.real
Xs_in= zin.imag

print("\n\n")
print("Valores de IMPEDANCIA SERIE de ENTRADA")
print("Rs_in=",Rs_in)
print("Xs_in=",Xs_in)

#Extrae la parte real e imaginaria de z_out

Rs_out= zout.real
Xs_out=zout.imag

print("\n")
print("Valores de IMPEDANCIA SERIE de SALIDA")
print("Rs_out=",Rs_out)
print("Xs_out=",Xs_out)

#Adaptacion de entrada

#Pasaje de impedancia serie a paralelo de z_in
Rp_in= Rs_in*(1+(Xs_in/Rs_in)**2)
Xp_in= Xs_in*(1+(Rs_in/Xs_in)**2)  #(Rs_in*Rp_in)/Xs_in

print("\n")
print("Valores de IMPEDANCIA PARALELO de ENTRADA")
print("Rp_in=",Rp_in)
print("Xp_in=",Xp_in)

#Cancelar reactancia
Xc_adpt_in= -Xp_in

C_in= 1/(2*np.pi*f*abs(Xc_adpt_in))

print("Xc_adapt_in= ", abs(Xc_adpt_in))
print("C_in=",C_in)

#lambda/4 compensacion de la parte real
zo_in= np.sqrt(50*Rp_in)

print("\n\nTrafo lambda/4 ENTRADA")
print("zo_in=",zo_in)

#Adaptacion de salida

print("\n\nTrafo lambda/4 SALIDA")
#lambda/4 compensacion de la parte real
zo_out= np.sqrt(50*Rs_out)
print("zo_out=",zo_out)

#usar el lambda/4 para pasar a modelo paralelo y cambiar el signo de la reactancia
Xl_adpt_out=(zo_out**2)/Xs_out

print("Xl_adpt_out=",Xl_adpt_out)

Xc_adapt_out= -Xl_adpt_out
print("Xc_adapt_out=",Xc_adapt_out)


#cancelar reactancia con el capacitor

C_out= 1/(2*np.pi*f*abs(Xc_adapt_out))

print("C_out=",C_out)
```

## Cálculo de microtiras

```python
import sympy as sym
import numpy as np
import cmath
from cmath import sqrt
import math as m



w_H, H, er, t, zo, Zo, lamda = sym.symbols('w_H, h, er, t, zo, Zo, lamda')

c= 300000000
H= 1.66*10**-3
er= 4.5
t= 0.04*10**-3

zo= 50

#Ecuaciones de Hammerstad
#Ecuacion No. 7
A=(zo/60)*((er+1)/2)**0.5+((er-1)/(er+1))*(0.23+(0.11/er))

#Ecuacion No. 8
B=(377*np.pi)/((2*zo)*(er)**0.5)

#w_H<2  Ecuación 5
w_H=(8*np.exp(A))/((np.exp(2*A))-2)
if(w_H >= 2):   #Si w/H resulta se mayor a 2 se utiliza la ecuación No. 6
    #w_H>2
    w_H=((2/np.pi)*(B-1-(m.log((2*B)-1, np.e))+(((er-1)/(2*er))*(m.log(B-1, np.e)+0.39-(0.61/er)))))

w= w_H*H

print("Resultados de la microtira de zo= ", zo)
print("A=",A)
print("B=",B)
print("w_H=",w_H)
print("w=",w)


#Calculo de la impedancia caracteristica
if (w_H)<=1:
    #Ecuacion No. 2 (Er efectivo)
    Eef=((er+1)/2)+(((er-1)/2)*(1/(1+((12*H)/w))**0.5))+(0.04*(1-(w/H))**2)

    #Ecuacion No. 1 (Impedancia caraterística de la microtira)
    Zo=(60/np.sqrt(Eef))*m.log((8*(H/w)+(w/(4*H))), np.e)    
else:
    #Ecuacion No. 4 (Er efectivo)
    Eef=((er+1)/2)+(((er-1)/2)*(1/((1+((12*H)/w))**0.5)))

    #Ecuacion No. 3 (Impedancia caraterística de la microtira)
    Zo=(((120*np.pi)/(np.sqrt(Eef)))/((w_H)+1.393+(0.667*m.log((w_H)+1.444, np.e))))

print("Eef=",Eef)
print("Zo=",Zo)

#Calculo de ancho efectivo
if (w_H)<=(1/2*np.pi):
    #Ecuacion No. 9
    we=w+((t/np.pi)*(1+m.log(4*np.pi*w/t, np.e)))
else:
    #Ecuacion No. 10
    we=w+((t/np.pi)*(1+m.log(2*H/t, np.e)))

print("we=",we)

#Calculo de la longitud de onda

lamda=(c/f)

lamda_eff=lamda/np.sqrt(Eef)

lamda_eff_4=lamda_eff/4

print("lamda=",lamda)
print("lamda_eff=",lamda_eff)
print("lamda_eff_4=",lamda_eff_4)
```
